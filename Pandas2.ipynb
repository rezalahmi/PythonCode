```python
def findComma(s):
    index = 1
    for i in s:
        if i != ',':
            index+=1
        else:
            break
    return index-1
```

```python
summer.Athlete.apply(lambda x : x[0:findComma(x)])
```

```python
summer.Athlete.map(lambda x : x[0:findComma(x)])
```
