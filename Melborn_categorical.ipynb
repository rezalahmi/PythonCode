```python
import pandas as pd
X = pd.read_csv(r'train.csv')
x_test = pd.read_csv(r'test.csv')
```

```python
## Remove rows with missing target, separate target from predictors
X.dropna(axis=0,inplace=True,subset=['SalePrice'])
y = X.SalePrice
X.drop(labels=['SalePrice'],axis=1,inplace=True)
```

```python
# To keep things simple, we'll drop columns with missing values
col_with_missing_values = [col for col in X.columns if X[col].isnull().any()]
X.drop(col_with_missing_values,axis=1,inplace=True)
x_test.drop(col_with_missing_values,axis=1,inplace=True)
```
