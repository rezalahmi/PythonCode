```python
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

```python
#read data set
X_full = pd.read_csv('train.csv')
```

```python
#remove data with missing target
X_full.dropna(axis=0,subset=['SalePrice'],inplace=True)
y = X_full['SalePrice']
X_full.drop(labels=['SalePrice'],axis=1,inplace=True)
```

now we split data to train set and test set

```python
X_train_full, X_valid_full, y_train, y_valid = train_test_split(X_full,y,
                                                               train_size=0.8,
                                                               test_size=0.2)
```

select categorical feature with low cardinality

```python
categorical_col = [cols for cols in X_full.columns
                  if X_full[cols].dtype == 'object' and
                  X_full[cols].nunique() < 10]
```

```python
numerical_col = [cols for cols in X_full.columns
                if X_full[cols].dtype in ['int64','float64']]
```

```python
X_train = X_train_full[categorical_col+numerical_col].copy()
X_valid = X_valid_full[categorical_col+numerical_col].copy()
```

Pipline model

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder,StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
#pre-processing for numerical data
numerical_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='median')),
    ('scaler',StandardScaler())])
#pre-processing for categorical data
categorical_transformer = Pipeline(steps=[
    ('imputer',SimpleImputer(strategy='most_frequent')),
    ('onehat',OneHotEncoder(handle_unknown='ignore'))])
# Bundle preprocessing for numerical and categorical data
preprocessor = ColumnTransformer(transformers=[
    ('num',numerical_transformer,numerical_col),
    ('cat',categorical_transformer,categorical_col)])
#create Model
myModel = RandomForestRegressor(n_estimators=300,random_state=0,criterion='mae',max_depth=13)
#make pipline for preprocessing and training
clf = Pipeline(steps=[
    ('preprocessor',preprocessor),
    ('model',myModel)])
#know train data
clf.fit(X_train,y_train)
#prediction
predict = clf.predict(X_valid)
print('mean absolute error:',mean_absolute_error(y_valid,predict))
```
