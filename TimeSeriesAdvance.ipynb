```python
yf.download('ETH-USD',start='2021-10-30',end='2021-10-31')
```

```python
stocks = pd.read_csv(r'stocks.csv',header=[0,1],index_col=[0],parse_dates=[0])
```

```python
stocks.describe()
```

```python
close = stocks.loc[:,'Close'].copy()
```

```python
close.head()
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

```python
close.plot(figsize=(15,8),fontsize=13)
plt.legend(fontsize=14)
plt.show()
```

```python
norm = close.div(close.iloc[0]).mul(100)
```

```python
norm.plot(figsize=(15,8),fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

```python
aapl = close.AAPL.copy().to_frame()
```

```python
aapl.head()
```

```python
aapl.shift(periods=1)
```

```python
aapl['lag1'] = aapl.shift(periods=1)
```

```python
aapl.head()
```

```python
aapl['diff'] = aapl.AAPL.sub(aapl.lag1)
```

```python
aapl.head()
```

```python
aapl['pct_change'] = aapl.AAPL.div(aapl.lag1).sub(1).mul(100)
```

```python
aapl.head()
```

```python
aapl[['diff','pct_change']].plot(figsize=(15,8),fontsize=10)
plt.legend(fontsize=12)
plt.savefig('test.jpeg')
plt.show()
```
